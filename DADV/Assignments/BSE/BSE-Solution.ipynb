# identify the top 10 gainers (%) across groups over the 5 years period. 
# Note: The gain is the closing price on the last day of the period, divided by the closing price on the first day of the period, minus 1.

```python
data = {}
for root,dirs,files in os.walk(os.curdir):
    for file in files:
        if file.endswith('.csv') and file.startswith("5"):
            name = os.path.splitext(file)[0]
            df = pd.read_csv(file,thousands=",")
            df["Date"] = pd.to_datetime(df["Date"])
            df.sort_values(by="Date",inplace=True)
            
            try :
                start = df["Close"].iloc[0]
            except Exception as e:
                start = df["Close Price"].iloc[0]
            
            try :
                end = df["Close"].iloc[-1]
            except Exception as e:
                end = df["Close Price"].iloc[-1]
            gain = (end/start) - 1
            data[name] = gain
```

```python
df = pd.DataFrame(data.items(),columns=["security","gain"])
df.sort_values(by="gain",inplace=True,ascending=False)
df.head(10)
```

```python
icici_bank = "ICICI_Bank"
df_icici_bank = pd.read_csv(icici_bank + ".csv",thousands=",")

stock_corr = pd.DataFrame()
for root,dirs,files in os.walk(os.curdir):
    for file in files:
        if file.endswith('.csv') and file.startswith("5"):
            name = os.path.splitext(file)[0]
            if name == icici_bank:
                continue
            df = pd.read_csv(file,thousands=",")
            cor = df_icici_bank["Total Turnover (Rs.)"].corr(df["Total Turnover (Rs.)"])
            try:
                stock_corr = stock_corr.append({'security' : name, 'correlation' : cor, 'name' : security_names[int(name)]},  
                ignore_index = True)
            except Exception as e:
                pass
            

stock_corr = stock_corr.sort_values(by="correlation",ascending=False)
stock_corr.head()
```

# Draw a correlation matrix of any 30 stocks on the Sensex. This is a matrix with both axes holding the stock names. The colour of each cell is the correlation between that pair of stocks: blue for -1 and green for 1.

```python
dfs = {}
stock_names = set()
def read_dfs():
    for root,dirs,files in os.walk(os.curdir):
        for file in files:
            if len(stock_names) == 30:
                return
            if file.endswith('.csv') and file.startswith("5"):
                name = os.path.splitext(file)[0]
                df = pd.read_csv(file,thousands=",")
                dfs[name] = df
                stock_names.add(name)
read_dfs()
```

```python
samp = pd.DataFrame(columns=stock_names,index=stock_names)
for n1,n2 in combinations(stock_names,2):
    df1 = dfs[n1]
    df2 = dfs[n2]
    c1 = df1["Total Turnover (Rs.)"].corr(df2["Total Turnover (Rs.)"])
    c2 = df2["Total Turnover (Rs.)"].corr(df1["Total Turnover (Rs.)"])
    samp.loc[n1,n2] = c1
    samp.loc[n2,n1] = c2
    samp.loc[n1,n1] = 1
    samp.loc[n2,n2] = 1
samp = samp.astype({name:"float" for name in stock_names})
```

```python
samp.style.background_gradient(cmap="brg")
```

# Draw a sparkline showing the data 23,80,92,62,98,7,9,56,19,68.


```python
import base64
import io


def spark_line(data, figsize=(4, 0.25), **kwags):
    data = list(data)

    fig, ax = plt.subplots(1, 1, figsize=figsize, **kwags)
    ax.plot(data)
    for k,v in ax.spines.items():
        v.set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])

    plt.plot(len(data) - 1, data[len(data) - 1], 'r.')

    ax.fill_between(range(len(data)), data, len(data)*[min(data)], alpha=0.1)

    img = io.BytesIO()
    plt.savefig(img, transparent=True, bbox_inches='tight')
    img.seek(0)
    plt.close()

    return base64.b64encode(img.read()).decode("UTF-8")
```

```python
values = [23,80,92,62,98,7,9,56,19,68]


with open("sparkline.html", "a") as file:
        file.write('<div><h1>Sparkline to the given values</h1></div>')
        file.write('<br>')
        file.write('<br>')
        file.write('<div><img src="data:image/png;base64,{}"/></div>'.format(spark_line(values)))
        file.write('<div></div>')
    
```

# For each stock, draw a sparkline showing last month's closing price trend

```python
closing_price_values = list()
for file in os.listdir():
    if file.endswith('.csv') and file.startswith("5"):
        filename = os.path.splitext(file)[0]
        if not filename.isdecimal() :
            continue
        df = pd.read_csv(file,thousands=",")
        try:
            closing_price_values.append(df["Close"][0:30].values)
        except Exception as e:
            closing_price_values.append(df["Close Price"][0:30].values)

with open("bse_closing_price.html", "a") as file:
    file.write('<div><h1>Sparkline to the Closing Price values</h1></div>')
    file.write('<br>')
    file.write('<br>')
    for value in closing_price_values:
        file.write('<div><img src="data:image/png;base64,{}"/></div>'.format(spark_line(value)))
        file.write('<div></div>')
```
